```python
#一次处理多个文档，没有replicated文档
#single chain with only tube

import pandas as pd
```

```python
#load file

fread = open('C:\\Users\\Think\\Documents\\polymer_python_research_code\\py_molecule_file_LAMMPS\\minsert.txt', 'r')
read1 = fread.readlines()
        
#strip, split and save data
read2 = []
for j in read1:
    first = j.strip('/n')
    second = first.split()
    read2.append(second)

read3 = []
for j in read2:
    if j != []:
        read3.append(j)

#Masses
ID_m = []
masses = []
mnumber = 120 #the total number of particles
mmass = 1.0 #mass of one particle
for j in range(1, mnumber+1):
    ID_m.append(j)
    masses.append(mmass)
    
#Types
types = []
ntype = 1 #types of particles
for j in range(1, mnumber+1):
    types.append(ntype)

#Coords
x_c = []
y_c = []
z_c = []
c_read = []
for j in range(9, len(read3)):
    if len(read3[j]) == 9:
        c_read.append(read3[j])
for j in range(0, len(c_read)):
    x_c.append(c_read[j][3])
    y_c.append(c_read[j][4])     
    z_c.append(c_read[j][5])

#Bonds
b_read = []
for j in range(9, len(read3)):
    if len(read3[j]) == 4:
        b_read.append(read3[j])
               
#write 
fwrite1 = open('minsert_ready.txt', 'w')
#write title
for j in range(0, 9, 1):
    a = str(read3[j]).replace('[','').replace(']','')
    a = a.replace("'",'').replace(',','') + '\n'
    b = '\n'
    fwrite1.write(a)
    fwrite1.write(b)
    
b = '\n'
fwrite1.write(b)

#write mass
c = 'Masses' + '\n'
fwrite1.write(c)
b = '\n'
fwrite1.write(b)
for j in range(0, len(masses)):
    d = str(ID_m[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + ' '
    e = str(masses[j]).replace('[','').replace(']','')
    e = e.replace("'",'').replace(',','') + '\n'
    fwrite1.write(d)
    fwrite1.write(e)
    
b = '\n'
fwrite1.write(b)

#write types
c = 'Types' + '\n'
fwrite1.write(c)
b = '\n'
fwrite1.write(b)
for j in range(0, len(types)):
    d = str(ID_m[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + ' '
    e = str(types[j]).replace('[','').replace(']','')
    e = e.replace("'",'').replace(',','') + '\n'
    fwrite1.write(d)
    fwrite1.write(e)

b = '\n'
fwrite1.write(b)

#write coords
c = 'Coords' + '\n'
fwrite1.write(c)
b = '\n'
fwrite1.write(b)
for j in range(0, len(ID_m)):
    d = str(ID_m[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + ' '
    e = str(x_c[j]).replace('[','').replace(']','')
    e = e.replace("'",'').replace(',','') + ' '
    f = str(y_c[j]).replace('[','').replace(']','')
    f = f.replace("'",'').replace(',','') + ' '
    g = str(z_c[j]).replace('[','').replace(']','')
    g = g.replace("'",'').replace(',','') + '\n'
    fwrite1.write(d)
    fwrite1.write(e)
    fwrite1.write(f)
    fwrite1.write(g)

b = '\n'
fwrite1.write(b)

#write bonds
c = 'Bonds' + '\n'
fwrite1.write(c)
b = '\n'
fwrite1.write(b)
for j in range(0, len(b_read)):
    d = str(b_read[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + '\n'
    fwrite1.write(d)

b = '\n'
fwrite1.write(b)

b = '\n'
fwrite1.write(b)
fwrite1.close()
```
