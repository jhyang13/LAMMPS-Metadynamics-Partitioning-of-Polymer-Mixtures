```python
#single chain or single cahin in mixture with metaD

import pandas as pd
import numpy as np
```

```python
#rearrange
#1_add chain number, time step
#load file
fread1 = open('C:\\Users\\Think\\Documents\\polymer_python_research_code\\py_enthalpy\\poly.lammpstrj', 'r')
read1 = fread1.readlines()
        
#strip, split and save data
read2 = []
for j in read1:
    first = j.strip('/n')
    second = first.split()
    read2.append(second)

#delete title in dump file
dump = []
length = 11 #列数****************
for j in range(0, len(read2), 1):
    if len(read2[j]) == length:
        dump.append(read2[j])

#add time step
timestep = []
idumpstep = 1100000 # timestep the dump start*******
dumpsegment = 1000  # dump every time step**********
chainlength = 120 # atoms number********************
for j in range(1, len(dump) // chainlength + 1, 1):
    for k in range(1, chainlength + 1, 1):
        timestep.append(idumpstep)
    idumpstep = idumpstep + dumpsegment

#write dump into a new file
fwrite1 = open('rearrange.txt', 'w')
a = 'ID TYPE x y z ix iy iz fx fy fz' + '\n'
fwrite1.write(a)
for j in range(0, len(dump), 1):
    c = str(dump[j]).replace('[','').replace(']','')
    c = c.replace("'",'').replace(',','') + '\n'
    fwrite1.write(c)
fwrite1.close()

#write timestep into a new file
fwrite1 = open('time.txt', 'w')
a = 'timestep' + '\n'
fwrite1.write(a)
for j in range(0, len(timestep), 1):
    a = str(timestep[j]).replace('[', '').replace(']', '')
    a = a.replace("'", '').replace(',', '') + '\n'
    fwrite1.write(a)
fwrite1.close()
```

```python
#pandas: particles in pore
#load files
dfl = pd.read_table('rearrange.txt', delim_whitespace=True)
dfl_t = pd.read_table('time.txt', delim_whitespace=True)

#calculate real coordination
xbox = 50 #the total length of simulation box*************
ybox = 50 #the total length of simulation box*************
zbox = 101 #the total length of simulation box************
dfl['realx'] = dfl.x + xbox*dfl.ix
dfl['realy'] = dfl.y + ybox*dfl.iy
dfl['realz'] = dfl.z + zbox*dfl.iz

#add timestep
dfl['timestep'] = dfl_t['timestep']

#select particles make contact with pore_sidewall
#dfl_contact_x_1 = dfl.loc[(dfl['realx']<= -0.5) | (dfl['realx']>= 0.5)]
#dfl_contact_y_1 = dfl_contact_x_1.loc[(dfl_contact_x_1['realy']<= -0.5) | (dfl_contact_x_1['realy']>= 0.5)]

#select particles make contact with pore sidewalls
bottom_side = -51 #下半截tube的长度***************
rc = 2.5 #截断半径********************************
bottom_wall = bottom_side + rc
dfl_contact_z_1 = dfl.loc[(dfl['realz'] <= 0) & (dfl['realz'] > bottom_wall)]

#select particles make contact with pore bottomwall
dfl_contact_z_2 = dfl.loc[(dfl['realz'] <= bottom_wall)]

#write files
dfl_contact_z_1.to_csv('contact_with_sidewall.txt', sep=' ', header=None, index=None)
dfl_contact_z_2.to_csv('contact_with_bottomwall.txt', sep=' ', header=None, index=None)
```

```python
#计算side wall的势
#load file
freadinital1 = open('contact_with_sidewall.txt', 'r')
initial = freadinital1.readlines()

#strip and split data
dump = []
for j in initial:
    first = j.strip('/n')
    second = first.split()
    dump.append(second)

#计算势
potential_value = 0
potential = []
timestep = []
EPS = 1.00 #吸引力的大小*****
rc = 2.5 #截断半径的大小*****
r_pore = 3 #tube的半径*******
for j in range(0,len(dump)):
    r1 = float(dump[j][11])
    r2 = float(dump[j][12])
    r3 = (r1**2 + r2**2)**0.5
    r = r_pore - r3
    if 0 < r <= 2.5:
        potential_value = EPS*(2/15*(1/r)**9 -(1/r)**3)
        potential.append(potential_value)
        timestep.append(dump[j][14])

#write into a new file
fwrite1 = open('potential_with_sidewall.txt', 'w')
a = 'timestep sidep' + '\n'
fwrite1.write(a)
for j in range(0, len(potential), 1):
    s = str(timestep[j]).replace('[','').replace(']','')
    s = s.replace("'",'').replace(',','') + ' '
    d = str(potential[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + '\n'
    fwrite1.write(s)
    fwrite1.write(d)
fwrite1.close()
```

```python
#计算bottom wall的势
#load file
freadinital1 = open('contact_with_bottomwall.txt', 'r')
initial = freadinital1.readlines()

#strip and split data
dump = []
for j in initial:
    first = j.strip('/n')
    second = first.split()
    dump.append(second)

#计算势
potential_value = 0
potential = []
timestep = []
EPS = 1.00 #吸引力的大小******
rc = 2.5 #截断半径************
r_pore = 3 #tube的半径********
bottomside = -51 #tube底的位置********
for j in range(0,len(dump)):
    r1 = float(dump[j][11])
    r2 = float(dump[j][12])
    r3 = float(dump[j][13])
    r4 = (r1**2 + r2**2)**0.5
    rxy = r_pore - r4
    rz = r3 - bottomside
    if rz <= rxy and 0 < rz <= 2.5:
        potential_value = EPS*(2/15*(1/rz)**9 -(1/rz)**3)
        potential.append(potential_value)
        timestep.append(dump[j][14])
    elif rz >= rxy and 0 < rxy <= 2.5:
        potential_value = EPS*(2/15*(1/rxy)**9 -(1/rxy)**3)
        potential.append(potential_value)
        timestep.append(dump[j][14])

#write into a new file
fwrite1 = open('potential_with_bottomwall.txt', 'w')
a = 'timestep bottomp' + '\n'
fwrite1.write(a)
for j in range(0, len(potential), 1):
    s = str(timestep[j]).replace('[','').replace(']','')
    s = s.replace("'",'').replace(',','') + ' '
    d = str(potential[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + '\n'
    fwrite1.write(s)
    fwrite1.write(d)
fwrite1.close()
```

```python
#pandas: 计算平均值
#load files
dfs = pd.read_table('potential_with_sidewall.txt', delim_whitespace=True)
dfb = pd.read_table('potential_with_bottomwall.txt', delim_whitespace=True)

#计算平均值
dfs_ave['s_timestep'] = dfs.groupby('timestep')['timestep'].mean()
dfs_ave['s_ave'] = dfs.groupby('timestep')['sidep'].mean()
dfs_re = pd.concat([dfs_ave['s_timestep'],dfs_ave['s_ave']], axis=1 )
dfs_re.to_csv('potential_with_sidewall_ave.txt', sep=' ', header=None, index=None)

dfb_ave['b_timestep'] = dfb.groupby('timestep')['timestep'].mean()
dfb_ave['b_ave'] = dfb.groupby('timestep')['bottomp'].mean()
dfb_re = pd.concat([dfb_ave['b_timestep'],dfb_ave['b_ave']], axis=1 )
dfb_re.to_csv('potential_with_bottomwall_ave.txt', sep=' ', header=None, index=None)
```

```python
#计算在每一个timestep下的平均值
#load file
freadinital1 = open('potential_with_sidewall_ave.txt', 'r')
initial1 = freadinital1.readlines()
#strip and split data
side = []
for j in initial1:
    first = j.strip('/n')
    second = first.split()
    side.append(second)

#load file
freadinital2 = open('potential_with_bottomwall_ave.txt', 'r')
initial2 = freadinital2.readlines()
#strip and split data
bottom = []
for j in initial2:
    first = j.strip('/n')
    second = first.split()
    bottom.append(second)    

#add time step
timestep = []
idumpstep = 1100000 #timestep the dump start*******
ldumpstep = 11100000 #timestep the dump last*******
dumpsegment = 1000  #dump every time step**********
for j in range(idumpstep, ldumpstep+1, dumpsegment):
    timestep.append(j)

#计算平均值
spotential = []
ave = 0
ave_re = []
for i in range(0, len(timestep)):
    for j in range(0, len(side)):
        if int(float(side[j][0])) == int(timestep[i]):
            spotential.append(float(side[j][1]))
    for k in range(0, len(bottom)):
        if int(float(bottom[k][0])) == int(timestep[i]):
            spotential.append(float(bottom[k][1]))
    if len(spotential) != 0:
        ave = sum(spotential) / len(spotential)
        ave_re.append(ave)
        spotential = []
    else:
        ave_re.append(0)
        
#write into a new file
fwrite1 = open('potential_re.txt', 'w')
for j in range(0, len(timestep), 1):
    s = str(timestep[j]).replace('[','').replace(']','')
    s = s.replace("'",'').replace(',','') + ' '
    d = str(ave_re[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + '\n'
    fwrite1.write(s)
    fwrite1.write(d)
fwrite1.close()
```
