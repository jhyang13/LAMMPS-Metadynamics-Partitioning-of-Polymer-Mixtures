```python
import pandas as pd
import numpy as np
```

```python
#rearrange
#load file
fread1 = open('C:\\Users\\jhyan\\Documents\\all\\projects\\csml_python_code\\py_energy_cv\\comz_pf_EPS0.40.txt', 'r')
read1 = fread1.readlines()
#strip, split and save data
read2 = []
for j in read1:
    first = j.strip('/n')
    second = first.split()
    read2.append(second)

#创建Z
zdirect_np = np.linspace(-31, 30, 610)
zdirect = []
for i in zdirect_np:
    zdirect.append(round(float(i),1))

#enthalpy延Z的方向
enthalpy = []
enthalpy_result = []
enthalpy_ave = 0
segment = 0.1
for i in range(0, len(zdirect), 1):
    for j in range(2, len(read2), 1):
        if float(read2[j][1]) < float(zdirect[i])+segment and float(read2[j][1]) >= float(zdirect[i]):
            enthalpy.append(float(read2[j][2]))            
    if len(enthalpy) == 0:
        enthalpy_result.append(0)
        enthalpy = []
    else:
        print(enthalpy)
        enthalpy_ave = sum(enthalpy) / len(enthalpy)
        enthalpy_result.append(float(enthalpy_ave))
        enthalpy = []
                
#write into a new file
fwrite1 = open('comz_energy_re.txt', 'w')
for j in range(0, len(enthalpy_result), 1):
    s = str(zdirect[j]).replace('[','').replace(']','')
    s = s.replace("'",'').replace(',','') + ' '
    d = str(enthalpy_result[j]).replace('[','').replace(']','')
    d = d.replace("'",'').replace(',','') + '\n'
    fwrite1.write(s)
    fwrite1.write(d)
fwrite1.close()
```
